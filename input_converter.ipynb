```python
import csv
import random
import math
```

```python
def from_csv_edges_to_adjacency_list(input_file, output_file):
    data = []
    with open(input_file, newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
    
    data.pop(0)
    
    n = 0
    for d in data:
        if (int(d[0]) > n):
            n = int(d[0])
        if (int(d[1]) > n):
            n = int(d[1]) 
    n += 1
    adj_list = [[] for _ in range(n)]
    for d in data:
        adj_list[int(d[0])].append(int(d[1]))
        adj_list[int(d[1])].append(int(d[0]))
        
    with open(output_file, 'w') as f:
        f.write(str(n+1))
        f.write("\n")
        for t in adj_list:
            f.write(str(len(t)))
            f.write(" ")
            for p in t:
                f.write(str(p))
                f.write(" ")
            f.write("/n")
```

```python
def from_edges_to_adjacency_list(input_file, output_file):
    data = []
    with open(input_file, 'r') as f:
        lines = f.readlines()
        for l in lines:
            data.append(l.split())
    
    n = 0
    for d in data:
        if (int(d[0]) > n):
            n = int(d[0])
        if (int(d[1]) > n):
            n = int(d[1]) 
    n += 1
    adj_list = [[] for _ in range(n)]
    for d in data:
        adj_list[int(d[0])].append(int(d[1]))
        adj_list[int(d[1])].append(int(d[0]))

        
    with open(output_file, 'w') as f:
        f.write(str(n))
        f.write('\n')
        for t in adj_list:
            f.write(str(len(t)))
            f.write(" ")
            for p in t:
                f.write(str(p))
                f.write(" ")
            f.write('\n')
```

```python
def sampling_from_edges_to_edges_with_n(input_file, output_file, c, e):
    data = []
    with open(input_file, 'r') as f:
        lines = f.readlines()
        for l in lines:
            data.append(l.split())
        f.close()
    
    n = 0
    for d in data:
        if (int(d[0]) > n):
            n = int(d[0])
        if (int(d[1]) > n):
            n = int(d[1]) 
    n += 1
    m = len(data)
    
    p = c * (1/e)**2 *math.log(n, 2) * n/m
    
    if (p > 1):
        print("Bad parameters for sampling")
        print("The probability of sampling is ", p)
        return
    
    sample = []
    new_n = 1 
    mapping = {}
    
    for d in data:
        t = random.random()
        if (t < p):
            if not((int(d[0]) in mapping.keys())):
                mapping[int(d[0])] = new_n
                new_n +=1
            if not((int(d[1]) in mapping.keys())):
                mapping[int(d[1])] = new_n
                new_n +=1
            sample.append([mapping[int(d[0])], mapping[int(d[1])]])
    
        
    with open(output_file, 'w') as f:
        f.write(str(new_n-1))
        f.write(' ')
        f.write(str(len(sample)))
        f.write('\n')
        for s in sample:
            f.write(str(s[0]))
            f.write(' ')
            f.write(str(s[1]))
            f.write('\n')
        f.close()
```
